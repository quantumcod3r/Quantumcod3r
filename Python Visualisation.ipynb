```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

#matplotlib practice

```python
data=[2,3,4,5,76,77,2,11,22,33,4,7]
plt.hist(data,bins=10,color="green",alpha=0.8)
```

    #Bar plot practice

```python
x=['A','B','C','D','E']
y=[2,5,8,12,17]
plt.bar(x,y,color="blue")
plt.xlabel('Data along horizontal')
plt.ylabel('Data along vertical')
plt.title('Bar Plot')
plt.savefig('bar.jpeg')
```

     line plottting

```python
data1=['A','B','C','D','E','F']
data2=[5,7,9,11,13,15]
data3=[9,13,17,21,25,29]
data4=[11,15,19,23,27,31]
style.use('ggplot')
plt.plot(data1,data2,'r',label='First data')
plt.plot(data1,data3,'g',label='Second data')
plt.plot(data1,data4,'b',label='Third data')
plt.xlabel("Data along Horizontal line")
plt.ylabel("Data along Vertical line")
plt.title("line Plotting")
plt.legend()
```

    Corona data in alphabetical order of states

```python
style.use('ggplot')
deceased=[0,198,1,12,77]
active=[58,8586,128,2895,2374]
recovered=[51,7313,66,6106,8020]
confirmed=[109,16097,195,9013,10471]
state=['Andaman and Nicobar','Andhra pradesh','Arunachal Pradesh','Assam','Bihar']
plt.plot(state,deceased,'r',label='Deceased')
plt.plot(state,active,'g',label='Active')
plt.plot(state,recovered,'b',label='Recovered')
plt.plot(state,confirmed,'Black',label='Confirmed')
plt.xlabel('Top five states COVID-19 data in alphabetical order')
plt.ylabel("Data as on 4th of july")
plt.xticks(rotation=45)
plt.legend()
```
